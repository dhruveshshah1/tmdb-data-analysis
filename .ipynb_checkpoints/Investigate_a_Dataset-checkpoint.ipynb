### Data Cleaning (Removing The Unused Information From The Dataset)
> #### Information That We Need To Delete Or Modify
1. We need to remove duplicate rows from the dataset
2. Remove the unused colums that are not needes in the analysis process.
3. Remove the movies which are having zero value of budget and revenue.

1.Dropping duplicate rows

```python
#drop all the duplicate rows
df.drop_duplicates(inplace=True)
print("After Removing Duplicate Values (Rows,Columns) : ",df.shape)
```

```python
# drop unneccessary variables
col=['budget_adj','revenue_adj','overview','imdb_id','homepage','tagline']
df.drop(col, axis=1, inplace=True)
df.head(1)
```

3. Drop theses rows which contain incorrect or inappropriate values.

```python
#drop rows where budget and revenue is 0
print("Rows With Zero Values In The Budget Column:",df[(df['budget']==0)].shape[0])
print("Rows With Zero Values In The Revenue Column:",df[(df['revenue']==0)].shape[0])
```

## Exploratory Data Analysis

### Research Question 1 (Find quantitative features )

### Research Question 2 ( Which Year Has The Highest Profit Rate?)

```python
#calculating profit
df['Profit'] = df['revenue'] - df['budget']
#Grouping by release year and plotting the graph of mean profit each year
df.groupby('release_year')['Profit'].mean().plot()
plt.title("Year Vs Average Profit",fontsize = 14)
plt.xlabel('Release year',fontsize = 12)
plt.ylabel('Average Profit',fontsize = 12)
```

According to the plot year 2002-03 has most profitable years And the profit was very low between the years 1960 and 1970

### Research Question 3  (Average runtime each year)

```python
#finding mean of average runtime and grouping by year
df.groupby('release_year')['runtime'].mean().plot()
plt.title("Year Vs Average Runtime",fontsize = 14)
plt.xlabel('Release year',fontsize = 12)
plt.ylabel('Average Profit',fontsize = 12)
```

 According to the plot movie duration is decreasing year to year and it's TRUE. Maybe because long duration movies get  boring. That's why the average runtime duraion of the movies are arround 100 Minutes. 

### Research Question 4  (Find the genre with the highest release)

```python
#funtion to split string and return count of each genre
def count_genre(x):
    #concatenate all the rows of the genres.
    data_plot = df[x].str.cat(sep = '|')
    data = pd.Series(data_plot.split('|'))
    #conts each of the genre and return.
    info = data.value_counts(ascending=False)
    return info

#call the function
total_genre_movies = count_genre('genres')
#plot a 'bar' plot
total_genre_movies.plot(kind= 'bar',figsize = (12,6))

#setup the title and the labels of the plot.
plt.title("Genre With Highest Release",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=12)
plt.ylabel("Genres",fontsize= 12)
```

According to the plot, Drama movies are the most released genre, followed by comedy. Western Genre is the least released genre, maybe because not many people like it.
